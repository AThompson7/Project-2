```python
# import dependencies
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

```python
# create path for calendar file and read
calendar_file = "Resources/calendar.csv"

calendar_df = pd.read_csv(calendar_file)
# print first 5 rows to check df
calendar_df.head()
```

```python
# exploring
# set to create a new df with columns 'listing_id', 'available', 'price', however, noticed that in the "available" column if value = f, the the value in the "price" column would be NaN, 
# therefore "available" column is not of much use as all NaN values pwill be removed. 

calendar_new_df = calendar_df[['listing_id', 'date','available', 'price']].copy()
calendar_new_df
```

```python
# drop all nan values
# clean_calendar_df = calendar_new_df.dropna()

# replace NaN values for $0 so that it can be used. 
calendar_new_df['price'] = calendar_new_df['price'] .fillna('$0')
clean_calendar_df = calendar_new_df

# place result in pandas df
# clean_calendar_df = pd.DataFrame(data= clean_calendar_df)

# remove duplicate if there are any
clean_calendar_df.drop_duplicates(inplace = True)

# print to review
clean_calendar_df.head()
```

```python
# create path for listings file and read
listings_file = "Resources/listings.csv"

listings_df = pd.read_csv(listings_file)
listings_df.head()
```

```python
# exploring
# set to create a new df with columns 'listing_id', 'available', 'price', however, noticed that in the "available" column if value = f, the the value in the "price" column would be NaN, 
# therefore "available" column is not of much use as all NaN values pwill be removed. 

calendar_new_df = calendar_df[['listing_id', 'date','available', 'price']].copy()
calendar_new_df
```

```python
# drop all nan values
clean_calendar_df = calendar_new_df.dropna()

# replace NaN values for $0 so that it can be used. 
calendar_new_df['price'] = calendar_new_df['price'] .fillna('$0')
clean_calendar_df = calendar_new_df

# place result in pandas df
# clean_calendar_df = pd.DataFrame(data= clean_calendar_df)

# remove duplicate if there are any
clean_calendar_df.drop_duplicates(inplace = True)

# print to review
clean_calendar_df.head()
```
